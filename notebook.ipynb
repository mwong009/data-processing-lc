```python
import os
import sys

import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_csv('sample_data.csv', header=None)
```

```python
df.columns=['instance_id', 'chunk_id', 'brake', 'gas', 'velocity', 'steering', 'lon_acc', 'lat_acc', 'image', 'category']
```

```python
df
```

```python
df.loc[df['category']==-1, 'category'] = 2
```

```python
df.brake = df.loc[:,['brake']]
```

```python
df.gas = df.loc[:,['gas']]
```

```python
df.lon_acc = df.loc[:,['lon_acc']]
df.lat_acc = df.loc[:,['lat_acc']]
```

```python
df.steering = df.loc[:,['steering']]
```

```python
scalar = preprocessing.StandardScaler().fit(df.loc[:,'brake':'lat_acc'])
```

```python
scaled_val = scalar.fit_transform(df.loc[:,'brake':'lat_acc'])
```

```python
df.loc[:,'brake':'lat_acc'] = scaled_val
```

```python
df
```

```python
df.std()
```

```python
df.mean()
```

```python
df_new = df.loc[:,'brake':'lat_acc'].merge(df.loc[:,['category']], left_index=True, right_index=True)
```

```python
train_data = df_new.sample(frac=1.)[:np.int(0.7 * df.shape[0])]
```

```python
train_data.iloc[:,-1].shape
```

```python
train_data = (train_data.iloc[:, :-1], train_data.iloc[:, [-1]])
```

```python
train_data
```

```python
import pickle
import gzip
```

```python
with gzip.open('dataset.pkl.gz', 'wb') as f:
    pickle.dump(train_data, f)
```
